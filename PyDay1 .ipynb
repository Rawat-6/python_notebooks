### Dictionaries in CSV

```python
# rather than getting data in list we can have a dictionary, having data as values and header names as keys so we 
# don't need to remember the indexes to extract data under particular header name

# classes DictReader and DictWriter as reader() & writer() functions

import csv
from pprint import pprint

with open('./us-500.csv') as file_point:
    dict_reader = csv.DictReader(file_point)  # instantiating a object of class DictReader  
    i = 0
    next(dict_reader)     # skipping 1st data row
    for record in dict_reader:
        if  i != 2:
            
            pprint(dict(record))                  #  Why record is list of tuples.....????
            i += 1   #print(type(record))
        else:
            break
        
# printing 1st two records

```

```python

date_string = '7/02/2018'
date_object = datetime.strptime(date_string,'%d/%m/%Y')
print(date_object)  # python prints in format YYYY-MM-DD
type(date_object)
```

```python
datetime_string = '7/06/2018 23:17'
date_obj = datetime.strptime(datetime_string,'%m/%d/%Y %H:%M')
print(date_obj)
```

```python
from datetime import timedelta   # for doing operations on date_time objects


print(date_obj)
date_obj = date_obj + timedelta(days=3)
print(date_obj)
date_obj = date_obj + timedelta(weeks=1)
print(date_obj)
date_obj = date_obj + timedelta(hours=1,minutes=3,seconds=4)
print(date_obj)
```

```python
print(date_obj-date_object)
print((date_obj-date_object).days)
print((date_obj-date_object).total_seconds())
#print((date_obj-date_object).hours)
```
